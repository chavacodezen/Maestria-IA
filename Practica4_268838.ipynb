*   Manipulación de matrices.
*   Crea dos matrices de 3x3 con números aleatorios entre 1 y 10. Realiza las siguientes operaciones:
1.   Suma ambas matrices.
2.   Resta la segunda matriz de la primera.
3.   Multiplica la primera matriz por 2.
4.   Realiza una multiplicación matricial entre las dos matrices.

```python
import numpy as np

# Matrices 3x3 (valores aleatorios entre 1 y 10)
matriz1 = np.random.randint(1, 11, size=(3, 3))
matriz2 = np.random.randint(1, 11, size=(3, 3))

print("Matriz 1:")
print(matriz1)

print("\nMatriz 2:")
print(matriz2)
```

```python
# Suma de ambas matrices (elemento a elemento)
suma = matriz1 + matriz2
print("\nSuma de matrices:")
print(suma)
```

```python
# Resta de la segunda matriz a la primera (elemento a elemento)
resta = matriz1 - matriz2
print("\nResta de matrices:")
print(resta)
```

```python
# Multiplicar la primera matriz por 2 (escalar)
multiplicacion_escalar = matriz1 * 2
print("\nMatriz 1 multiplicada por 2:")
print(multiplicacion_escalar)
```

```python
# Multiplicación matricial (producto punto)
multiplicacion_matricial = np.dot(matriz1, matriz2)

print("\nMultiplicación matricial:")
print(multiplicacion_matricial)
```
