Actividad:
*   Creación de un arreglo y operaciones básicas.
*   Crea un arreglo unidimensional de 20 elementos con valores aleatorios entre 1 y 100. Luego, realiza las siguientes operaciones:
1.   Encuentra el valor máximo del arreglo.
2.   Encuentra el valor mínimo del arreglo.
3.   Calcula la media del arreglo.
4.   Suma 10 a cada elemento del arreglo.

```python
import numpy as np

# Arreglo unidimensional de 20 elementos (valores aleatorios entre 1 y 100)
arreglo = np.random.randint(1, 101, size=20)

print("Arreglo original:")
print(arreglo)
```

```python
# Encontrar el valor maximo del arreglo
valor_maximo = arreglo.max()
print("\nValor maximo:", valor_maximo)
```

```python
# Encontrar el valor minimo del arreglo
valor_minimo = arreglo.min()
print("Valor minimo:", valor_minimo)
```

```python
# Calcular la media del arreglo
media = arreglo.mean()
print("Media del arreglo:", media)
```

```python
# Sumar 10 a cada elemento del arreglo
arreglo_mas_10 = arreglo + 10
print("\nArreglo después de sumar 10 a cada elemento:")
print(arreglo_mas_10)
```
