Actividad:
*    Uso de máscaras booleanas.
*   Crea un arreglo unidimensional con 15 números aleatorios entre 1 y 50. Luego:
1.   Selecciona los elementos mayores que 25 utilizando una máscara booleana.
2.   Cuenta cuántos elementos son mayores que 25.

```python
import numpy as np

# Arreglo unidimensional de 15 elementos (valores aleatorios entre 1 y 50)
arreglo = np.random.randint(1, 51, size=15)

print("Arreglo original:")
print(arreglo)
```

```python
# Mascara booleana (True si el valor es mayor que 25)
mascara = arreglo > 25

print("\nMascara booleana (valores > 25):")
print(mascara)
```

```python
# Elementos mayores que 25 usando la mascara
mayores_25 = arreglo[mascara]

print("\nElementos mayores que 25:")
print(mayores_25)
```

```python
# Elementos mayores que 25
cantidad_mayores_25 = np.sum(mascara)

print("\nElementos mayores que 25:")
print(cantidad_mayores_25)
```
