Actividad:
*   Tu código debe extraer únicamente los datos numéricos de las líneas que contienen la cadena X-DSPAM-Confidence en el archivo mbox.txt.
*   Para extraer los datos númericos usa las intrucciones vistas en el laboratorio 1 (startswith). Aquí no se usan expresiones regulares
*   Los datos numéricos los deberás guardar en una lista y posteriormente mostrar los resultados de cada medición encontrada.


```python
# Mount Drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Lista de valores numericos
valores = []

with open('/content/drive/MyDrive/ClassFiles/mbox.txt', 'r') as archivo:
    for linea in archivo:
        # Eliminamos espacios en blanco
        linea = linea.rstrip()

        # Filtrar solo las líneas con X-DSPAM-Confidence
        if linea.startswith('X-DSPAM-Confidence:'):
            numero = float(linea.split(':')[1])
            valores.append(numero)
```

```python
# Cantidad de datos extraidos
cantidad = len(valores)
print('Cantidad de datos extraídos:', cantidad)
```

```python
# Sumatoria de los datos
suma = 0
for valor in valores:
    suma = suma + valor
print('Sumatoria de los datos:', suma)
```

```python
# Promedio de los datos
promedio = suma / cantidad
print('Promedio de los datos:', promedio)
```

```python
# Varianza de los datos
varianza = 0
for valor in valores:
    varianza = varianza + (valor - promedio) ** 2

varianza = varianza / cantidad
print('Varianza de los datos:', varianza)
```
